# Converting the columns to appropriate data types.

```python
df['Date']=pd.to_datetime(df['Date'], errors="coerce", dayfirst=True)
num_cols = ['Open','Close','High','Low','Volume']
for col in num_cols:
    df[col]=(df[col].astype(str).str.replace(r"[^\d.]","",regex=True).astype(float))
```

```python
df.describe()
```

# Daily Percentage Change on Close to calculate daily returns

```python
df['daily_returns']=df['Close'].pct_change()
```

```python
df[['Date','Close','daily_returns']]
```

```python
df.sort_values(by='daily_returns',ascending=False).head(5)
```

```python
df['SMA_20']=df['Close'].rolling(window=20).mean()
df['SMA_50']=df['Close'].rolling(window=50).mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')

plt.figure(figsize=(12,5))
plt.plot(df.index,df['Close'],label='Closing Price', color='blue')
plt.plot(df.index,df['SMA_20'],label='20 Day Moving Average', color='orange')
plt.plot(df.index,df['SMA_50'],label='50 Day Moving Average', color='green')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('BHP stock Prices and Moving Averages(2024)')
plt.legend()

plt.savefig(r'C:\Users\Bhupal\OneDrive\Desktop\BHP_stock_price_project\bhp_stock_prices_moving_avg.png',dpi=300)
plt.show()
```

```python
plt.figure(figsize=(12,5))

sns.histplot(df['Volume'],bins=30, kde=True, color='purple' )
plt.xlabel('Trading volume')
plt.ylabel('Frequency')
plt.savefig(r'C:\Users\Bhupal\OneDrive\Desktop\BHP_stock_price_project\trading_volume.png',dpi=300)
```

# Average Trading Volume by Day of the Week

```python
df['Day_of_Week']=df['Date'].dt.day_name()
```

```python
day_order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
avg_vol_by_day=df.groupby('Day_of_Week')['Volume'].mean()
avg_vol_by_day=avg_vol_by_day.reindex(day_order)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df.groupby('Day_of_Week')['Volume'].mean().index, y=df.groupby('Day_of_Week')['Volume'].mean(), order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'], palette='viridis')

plt.xlabel('Day of Week')
plt.ylabel('Average Trading Volume')
plt.legend()

plt.savefig(r'C:\Users\Bhupal\OneDrive\Desktop\BHP_stock_price_project\avg_vol_by_day.png',dpi=300)
plt.show()
```

```python
df['Price_diff']=df['High']-df['Low']

plt.figure(figsize=(12,6))
sns.boxplot(x=df['Day_of_Week'],y=df['Price_diff'], order=['Monday','Tuesday','Wednesday','Thursday','Friday'],palette='coolwarm')

plt.xlabel('Day of Week')
plt.ylabel('Price Difference between High and Low')
plt.title('Stock Price Fluctuation by day of the week')

plt.savefig(r'C:\Users\Bhupal\OneDrive\Desktop\BHP_stock_price_project\stock_price_fluctuation.png',dpi=300)
plt.show()
```
